```python
PA_fin_df['pos_neg'] = 0

# 4,5점일때 1 / 1,2,3점일때 0 값 부여
PA_fin_df.loc[PA_fin_df['rating'] > 3, 'pos_neg'] = 1
def combine_2rd_columns(col_1, col_2):
    result = col_1
    if not pd.isna(col_2):
        result += " " + str(col_2)
    return result

PA_fin_df["title_text"] = PA_fin_df.apply(lambda x: combine_2rd_columns(x['title'], x['text']), axis=1)
PA_fin_df.to_csv('project.csv')

PA_fin_df
```
